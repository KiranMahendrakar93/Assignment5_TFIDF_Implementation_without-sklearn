   <h4><strong>What does tf-idf mean?</strong></h4>

   <p>    
Tf-idf stands for <em>term frequency-inverse document frequency</em>, and the tf-idf weight is a weight often used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. Variations of the tf-idf weighting scheme are often used by search engines as a central tool in scoring and ranking a document's relevance given a user query.
</p>
    
   <p>
One of the simplest ranking functions is computed by summing the tf-idf for each query term; many more sophisticated ranking functions are variants of this simple model.
</p>
    
   <p>
Tf-idf can be successfully used for stop-words filtering in various subject fields including text summarization and classification.
</p>
    
</font>

<font face='georgia'>
    <h4><strong>How to Compute:</strong></h4>

Typically, the tf-idf weight is composed by two terms: the first computes the normalized Term Frequency (TF), aka. the number of times a word appears in a document, divided by the total number of words in that document; the second term is the Inverse Document Frequency (IDF), computed as the logarithm of the number of the documents in the corpus divided by the number of documents where the specific term appears.

 <ul>
    <li>
<strong>TF:</strong> Term Frequency, which measures how frequently a term occurs in a document. Since every document is different in length, it is possible that a term would appear much more times in long documents than shorter ones. Thus, the term frequency is often divided by the document length (aka. the total number of terms in the document) as a way of normalization: <br>

$TF(t) = \frac{\text{Number of times term t appears in a document}}{\text{Total number of terms in the document}}.$
</li>
<li>
<strong>IDF:</strong> Inverse Document Frequency, which measures how important a term is. While computing TF, all terms are considered equally important. However it is known that certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scale up the rare ones, by computing the following: <br>

$IDF(t) = \log_{e}\frac{\text{Total  number of documents}} {\text{Number of documents with term t in it}}.$
for numerical stabiltiy we will be changing this formula little bit
$IDF(t) = \log_{e}\frac{\text{Total  number of documents}} {\text{Number of documents with term t in it}+1}.$
</li>
</ul>

<br>
<h4><strong>Example</strong></h4>
<p>

Consider a document containing 100 words wherein the word cat appears 3 times. The term frequency (i.e., tf) for cat is then (3 / 100) = 0.03. Now, assume we have 10 million documents and the word cat appears in one thousand of these. Then, the inverse document frequency (i.e., idf) is calculated as log(10,000,000 / 1,000) = 4. Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12.
</p>

### Your custom implementation

```python
# Write your code here.
# Make sure its well documented and readble with appropriate comments.
# Compare your results with the above sklearn tfidf vectorizer
# You are not supposed to use any other library apart from the ones given below

from collections import Counter
from tqdm import tqdm
from scipy.sparse import csr_matrix
import math
import operator
from sklearn.preprocessing import normalize
import numpy
```

```python
def fit_1(dataset):
    '''
    This function fits the given dataset by creating vocabulary of unique words having length more than 1 in the entire dataset
    '''

    unique_words = set()       # Initialize empty set of unique_words
    if isinstance(dataset, list):   # Check if dataset passed is in list format
        for row in dataset:         # For each document/review in dataset
            for word in row.split(" "):    # For each word in document/review. # splits document/review by space
                if len(word) >= 2:             # Filter single letter words
                    unique_words.add(word)     # Adding word to unique_words set
        unique_words = sorted(list(unique_words))     # covert to list and sort it
        vocab = {j:i for i,j in enumerate(unique_words)}  # Create final vocab with unique words as features and values as their column index

        return vocab

    else:
        print('Please pass the list of sentences')
```

```python
def IDF(dataset, vocab):
    '''
    This function computes the inverse document frequency of each word in the vocabulary
    '''

    idf_dict = {}             # Initialize an empty dictionary
    num_of_docs = len(dataset)
    for word in vocab.keys():  # For each word in vocabulary
        count = 0
        for row in dataset:    # For each row in dataset
            if word in row.split(' '):
                count += 1     # count increases by one if word is present in document

        idf = 1 + (math.log((1 + num_of_docs) / (1 + count)))
        idf_dict[word] = idf    # appends idf word and its idf value to the dictionary

    return idf_dict
```

```python
def transform_1(dataset, vocab):
    '''
    This function returns sparse matrix of tf-idf values for all unique words having length more than 1
    '''

    rows = []   # Initialize empty list for row_index of sparse matrix
    cols = []   # Initialize empty list for col_index of sparse matrix
    values = [] # Initialize empty list for non-zero values of sparse matrix

    if isinstance(dataset, list):     # Check if dataset is in list format
        num_of_docs = len(dataset)
        idf_dict = IDF(dataset, vocab)

        for idx, row in enumerate(dataset):     # For each document/review
            word_freq = dict(Counter(row.split(' ')))   # Get each word frequency in document/review
            num_of_words = len(row.split(" "))          # number of words in the document/review

            for word, freq in word_freq.items():      # For each word in word_freq dictionary
                if word in vocab.keys():              # Filter non-vocab words
                    # Term-Freq = Number of times term t appears in a document / Total number of terms in the document
                    TF = freq / num_of_words

                    IDF_word = idf_dict[word]

                    # Store the tf_idf values in sparse representation
                    col_idx = vocab[word]
                    rows.append(idx)
                    cols.append(col_idx)
                    values.append(TF*IDF_word)

        tfidf_matrix = csr_matrix((values, (rows, cols)), shape = (len(dataset), len(vocab.keys())))
        return normalize(tfidf_matrix, norm= 'l2', axis=1)  # Normalize the row vector

    else:
        print('Please pass list of sentences')
```

```python
def fit_2(dataset, max_features = None):
    '''
    This function fits the given dataset by creating vocabulary of unique words with length more than 1
    and top words based on their idf values if number of maximum features is passed
    '''

    unique_words = set()       # Initialize empty set of unique_words
    if isinstance(dataset, list):   # Check if dataset passed is in list format
        for row in dataset:         # For each document in dataset
            for word in row.split(" "):  # For each word in document # splits document by space
                if len(word) >= 2:             # Filter single letter words
                    unique_words.add(word)     # Adding word to unique_words set
        unique_words = sorted(list(unique_words))     # covert to list and sort it

        # Create final vocab with unique words as features and values as their column index
        vocab = {j:i for i,j in enumerate(unique_words)}
        top_features = None

        if max_features:
            idf_dict = IDF(dataset, vocab)     # Creating dictionary of idf values for each unique word
            # print(idf_dict)
            top_features = dict(sorted(idf_dict.items(), key = lambda x: x[1], reverse = True)[:max_features])
            vocab = {j:i for i,j in enumerate(list(top_features.keys()))}

        return vocab, top_features

    else:
        print('Please pass the list of sentences')
```
